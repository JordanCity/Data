```python
def cleanhtml(raw_html):
  import re
  clean = re.compile('<.*?>|&([a-z0-9]+|#[0-9]{1,6}|#x[0-9a-f]{1,6});') 
  cleantext = re.sub(clean, '', raw_html)
  return cleantext
```

```python
import pandas as pd
df_train=pd.read_csv('/content/cleaned_fashion_data.csv')
df_train.info()
```

```python
Df_train_final=df_train.loc[:, ['price', 'rating_count', 'avg_rating']]
Df_train_final.head()
```

```python
x_train= Df_train_final.loc[:, ['rating_count', 'avg_rating']]
x_train
```

```python
y_train= Df_train_final.loc[:, ['price']]
y_train
```

```python
Df_test_final=df_train.iloc[0:2000]
Df_test_final
```

```python
x_test= Df_test_final.loc[:, ['rating_count', 'avg_rating']]
x_test
```

```python
y_test= Df_test_final.loc[:, ['price']]
y_test
```

```python
import pandas as pd
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

reg= linear_model.LinearRegression()
model= reg.fit(x_train, y_train)
```

```python
predict_y= model.predict(x_test)
predict_y
```

```python
model.score(x_test, y_test)
```

```python
plt.figure(figsize=(18,10))
x_axis=range(len(y_test))
plt.scatter(x_axis, y_test, label='original')
plt.scatter(x_axis, predict_y, label='predicted')
plt.title('Actual and predicted data')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
import numpy as np
print("Mean squared error", metrics.mean_squared_error(y_test, predict_y))
print('Mean absolute error', metrics.mean_absolute_error(y_test, predict_y))
print('Root mean squared error', np.sqrt(metrics.mean_squared_error(y_test, predict_y)))
```
